```python
# Dependencies
import pandas as pd
```

```python
# Read with pandas
csv_path = "Resources/2016-FCC-New-Coders-Survey-Data.csv"


# Import the books.csv file as a DataFrame
bootcamp_df = pd.read_csv(csv_path, encoding ='iso-8859-1', low_memory=False)
bootcamp_df

```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
reduced_df = bootcamp_df.iloc [:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
reduced_df
```

```python
reduced_df.dtypes
```

```python
reduced_recode_df = reduced_df.replace({0.0: "No", 1.0: "Yes"})
reduced_recode_df

reduced_recode_df
```

```python
total_surveyed = len(reduced_recode_df)
total_surveyed
```

```python
attendee_status = reduced_recode_df["AttendedBootcamp"].value_counts()
attendee_status
```

```python
attendee_percent = reduced_recode_df["AttendedBootcamp"].value_counts(normalize =True)
print (attendee_percent *100)
```

```python
attendees = reduced_recode_df.loc[reduced_recode_df["AttendedBootcamp"] == "Yes"]
attendees.count()
```

```python
average_age = attendees["Age"].mean()
average_age
```

```python
degree_status = attendees["SchoolDegree"].count()
degree_status
```

```python
degree_percent = attendees['SchoolDegree'].value_counts(normalize =True)
print (degree_percent *100)
```

```python
##RECODE GENDER
attendees = attendees.replace(
    {"male": "Male", "female": "Female", "genderqueer": "Non-Binary", "trans": "Non-Binary", "agender" : "Non-Binary", "Non-binary" : "Non-Binary"})
attendees
```

```python
gender_count = attendees['Gender'].value_counts()
gender_count
```

```python
gender_percent = attendees['Gender'].value_counts(normalize =True)
print (gender_percent *100)
```

```python
average_salary = attendees["BootcampPostSalary"].mean()
average_salary
```
